```python
#Insertion sort
nums=[3,4,5,6,8,9,10,7,1]
n=len(nums)-1
for i in range(1,n+1):
    key=nums[i]
    j=i-1
    while j>=0 and nums[j]>key:
        nums[j+1]=nums[j]
        j=-1
    nums[j+1]=key
print(nums)
```

```python
arr1 = [1,1,2,3,4]
arr2 = [1,5,6]
n1 = len(arr1) - 1
n2 = len(arr2) - 1
result = []
i = j = 0
# Merge both arrays
while i <= n1 and j <= n2:
    if arr1[i] <= arr2[j]:
        if len(result) == 0 or result[-1] != arr1[i]:
            result.append(arr1[i])
        i += 1
    else:
        if len(result) == 0 or result[-1] != arr2[j]:
            result.append(arr2[j])
        j += 1

# Remaining elements of arr1
while i <= n1:
    if result[-1] != arr1[i]:
        result.append(arr1[i])
    i += 1

# Remaining elements of arr2
while j <= n2:
    if result[-1] != arr2[j]:
        result.append(arr2[j])
    j += 1

print(result)
```

```python
#find the maximum gap
def maximumGap(nums):
    if len(nums)<2:
        return 0
    nums.sort()
    max_gap=0
    i=1
    while i<len(nums):
        gap=nums[i]-nums[i-1]
        if gap>max_gap:
            max_gap=gap
        i+=1
    return max_gap
nums=[3,6,9,21]
maximumGap(nums)
```

```python
nums = [7, 8, 9, 1, 2, 3, 4]
n = len(nums)
l, h = 0, n - 1

# Change: Initialize with positive infinity or the first element
mini = float('inf') 

while l <= h:
    mid = (l + h) // 2
    
    # If the right side is sorted, the minimum could be mid, 
    # or it is somewhere to the left.
    if nums[mid] <= nums[h]:
        mini = min(mini, nums[mid])
        h = mid - 1
    
    # If the left side is sorted, the minimum must be on the right side
    # because the "pivot" (drop point) is there.
    else:
        mini = min(mini, nums[mid])
        l = mid + 1

print(mini)
```
