```python
num=int(input("Enter the number:"))
def is_prime(num):
    if num<=1:
        return False
    for i in range(2,num):
        if num % i==0:
            return False
    return True
total=0
prime_number=[]
for i in range(2,num+1):
    if is_prime(i):
        prime_number.append(i)
        total=total+i
print("Prime Numbers:",prime_number)
print("Sum of prime numbers:",total)
        
```

```python
s = 'aabbccddeeee'
count = 1
result = ''

for i in range(1, len(s) + 1):
    if i < len(s) and s[i] == s[i-1]:
        count += 1
    else:
        result += s[i-1] + str(count)
        count = 1

print(result)
```
