```python
#shallow copy
lst=[1,2]
lst2=lst.copy()#seprated copy created
lst2.append(3)
print(lst)
print(lst2)
print("shallow copy with nested lists")
lst1=[[1,2],[2,3],[4,5]]
lst2=lst1.copy()
lst2[1][1]=4#reflect original and copy strings
print(lst2)
print(lst1)
lst2.insert(1, [9, 9])#only changes will be done on the copy str          
print(lst2)
print(lst1)
#value will be change only second list not first
lst2[1].pop()
lst2[1].append(4)
lst2[1].insert(1,99)
print(lst1)
print(lst2)
```

```python
#deep copy
import copy
lst=[1,2]
lst2=copy.deepcopy(lst)
lst2.append(4)
print(lst)
print(lst2)

lst1=[[1,2],[3,4],[5,6]]
lst2=copy.deepcopy(lst1)
lst2[1][0]=99#in deep copy changes will not reflect with both copy
print(lst1)
print(lst2)
```

```python
#abstract methed
from abc import ABC ,abstractmethod
class Vehicle(ABC):
    def fuel(self):
        print("Fuel is needed") 
    @abstractmethod
    def speed(self):
        pass
class Bike(Vehicle):
    def speed(self):
        print("bike speed is 80 km/h")
obj=Bike()
obj.fuel()
obj.speed()
```

```python
# all() with empty list
print(all([]))     # True

# Reason:
# all() ka matlab → "kya sab elements True hain?"
# Empty list me koi element hi nahi hota
# So logically: koi bhi False nahi mila
# ⇒ condition satisfied
# This is called "vacuous truth"


# any() with empty list
print(any([]))     # False

# Reason:
# any() ka matlab → "kya koi ek element True hai?"
# Empty list me koi element hi nahi hota
# So True milne ka chance hi nahi
# ⇒ False
```

```python
num=int(input("Enter the number:"))
def prime(num):
    if num<=1:
        return False
    for i in range(2,num):
        if num%i==0:
            return False
    return True
prime(num)
total=0
prime_number=[]
for i in range(2,num+1):
    if prime(i):
        total=total+i
        prime_number.append(i)
print("Prime numbers",prime_number)
print("Sum of the primes numbers :",total)

```

```python
str='madam'
def palindrome(str):
    f=0
    l=len(str)-1
    while f<=l:
        if str[f]!=str[l]:
            return False
        else:
            f=f+1
            l=l-1
    return True

palindrome(str)
```

```python
s = 'aabbccddeeee'
result = ''
count = 1

for i in range(1, len(s) + 1):
    
    if i < len(s) and s[i] == s[i-1]:
        count += 1
    else:
        result += s[i-1] + str(count)
        count = 1

print(result)
```
