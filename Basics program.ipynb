<h1>Reverse of number</h1>

```python
num = int(input("Enter the number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10   # âœ… Corrected here

print("Reverse of number:", rev)
```

<h1>Palindrome of Number</h1>

```python
num = int(input("Enter the number: "))
temp = num   # store original number
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print("It is a Palindrome number")
else:
    print("It is NOT a Palindrome number")
```

<h1>Find maximum and minimume from the list</h1>

```python
lst=[32,43,435,456,56]
print(lst)
n=len(lst)
print(n)
for i in range(0,n):
    for j in range(i+1,n):
        if lst[i] > lst[j]:
            temp=lst[i]
            lst[i]=lst[j]
            lst[j]=temp
print("Sorted list:",lst)
print("minimum value",lst[0])
print("maximum value",lst[n-1])
print("second minimum ",lst[1])
print("second largest",lst[n-2])
```
