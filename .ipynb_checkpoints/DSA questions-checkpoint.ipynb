```python
#print 1 to 5 using recurison 
#This is the head recrusion .first job then funciton call , it called the head recurion
def dis(i,n):
    if i>n:
        return
    else:
        print(i)
        dis(i+1,n)
dis(1,5)
```

```python
#print n to 1 using recurison 
#This is the tail  recrusion .first funtion call then job , it called the tail recursion or backtracking
def dis(i,n):
    if i>n:
        return

    dis(i+1,n)
    print(i)
    
dis(1,5)
```

```python
word = "madakkjfd"
l = 0
r = len(word) - 1

def check_palindrome(word, l, r):
    # Base condition: indexes crossed
    if l >= r:
        return True

    # Characters do not match
    if word[l] != word[r]:
        return False

    # Check remaining middle part
    return check_palindrome(word, l + 1, r - 1)

# Function call
result=check_palindrome(word, l, r)
if result:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
#using recursion
def fib(n):
    if n==0 or n==1:
        return n

    else:
        return fib(n-1)+fib(n-2)
fib(9)
```

<h1>Selection Sort</h1>

```python
arr=[6,4,0,544,8,9,10]
print("Before sorting array in ascending order",arr)
l=len(arr)
print(l)
for i in range(0,l):
    for j in range(i+1,l):
        if arr[i]> arr[j]:
            arr[i],arr[j]=arr[j],arr[i]
           
print("After sorting array in ascending order",arr)
        
```

```python
arr = [1, 2, 3, 4]
print("Before sorting array in descending order",arr)
l=len(arr)-1
print(l)
for i in range(0,l+1):
    for j in range(i+1,i):
        if arr[i] < arr[j]:
            temp=arr[i]
            arr[i]=arr[j]
            arr[j]=temp

print("After sorting array in descendin order",arr)
        
```

<h3>Bubble Sort</h3>
<p>
Bubble Sort (short and simple)

Bubble Sort compares adjacent elements

If the order is wrong, it swaps them

After each pass, the largest element moves to the end

Process repeats until the list is sorted

Time Complexity

O(nÂ²)
</p>

```python
arr=[5,1,6,8,2,4,9]
l=len(arr)-1
print(l)
for i in range(l-1,-1,-1):
    for j in range(0,i+1):#j will be execute till i 
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
            
            
```

<h1>Merge Sort</h1>

```python
lst = [5, 4, 3, 2, 1]

def merge_sort(lst):
    # Base case
    if len(lst) <= 1:
        return lst

    # Split the list
    mid = len(lst) // 2
    left_lst = lst[:mid]
    right_lst = lst[mid:]

    # Recursive sorting
    left_sort = merge_sort(left_lst)
    right_sort = merge_sort(right_lst)

    # Merge step
    return merge(left_sort, right_sort)


def merge(left, right):
    result = []
    i = 0
    j = 0

    # Compare and merge
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Add remaining elements
    while i < len(left):
        result.append(left[i])
        i += 1

    while j < len(right):
        result.append(right[j])
        j += 1

    return result


sorted_lst = merge_sort(lst)
print(sorted_lst)
```
