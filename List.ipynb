Selection Sort – Definition

Selection Sort is a simple sorting algorithm that works by repeatedly finding the smallest element from the unsorted part and placing it at its correct position.

```python
#Selection Sort
arr=[6,4,0,544,8,9,10,2,23,434]
l=len(arr)-1
for i in range(0,l+1):
    for j in  range(i,l+1):
        if arr[i]>arr[j]:
            arr[i],arr[j]=arr[j],arr[i]
print("Sorted Array/n",arr)
```

```python
Bubble Sort Optimization – Definition
🔹 Explanation:
In every pass, the largest element moves to its correct position at the end
That portion becomes sorted and fixed
So, we reduce unnecessary comparisons by ignoring the sorted part
```

```python
#Bubble Sort
arr=[6,4,0,544,8,9,10,2,23,434]
l=len(arr)-1
for i in range(l,-1,-1):
    for j in range(0,i):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
```

Insertion Sort – Definition

Insertion Sort is a simple sorting algorithm that works by taking one element at a time and placing it in its correct position in the already sorted part of the list.

```python
#Insertation sort
arr=[3,5,6,4,8,9,10,7,1]
l=len(arr)-1
for i in range(1,l+1):
    key=arr[i]
    j=i-1
    while j>=0 and arr[j]>key:
        arr[j+1]=arr[j]
        j-=1
    arr[j+1]=key
print(arr)
```

```python
#Set Matrix Zeros -Brute solution

matrix = [[2,5,1],[3,0,2],[1,1,3]]

rows = len(matrix)
cols = len(matrix[0])

# Mark row and column as -inf
def setInfinity(matrix, row, col):

     # mark entire column
    for i in range(rows):
        if matrix[i][col] != 0:
            matrix[i][col] = float('-inf')
            
    # mark entire row
    for j in range(cols):
        if matrix[row][j] != 0:
            matrix[row][j] = float('-inf')
    
   

def setzeromatrix(matrix, rows, cols):
    # Step 1: mark -inf
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                setInfinity(matrix, i, j)

    # Step 2: convert -inf to 0
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == float('-inf'):
                matrix[i][j] = 0

setzeromatrix(matrix, rows, cols)

print(matrix)
```

```python
#find lowerbound
nums=[1,1,1,2,3,5,6,7,7,9,12,12,13]
target=16
def lowerbound(nums,target):
    l=0
    h=len(nums)-1
    lb=len(nums)+1#if target is not found
    while l<=h:
        mid=(l+h)//2
        if nums[mid] >= target:
            lb=mid
            h=mid-1
        else:
            l=mid+1
    return mid
lowerbound(nums,target)
```

```python
# Search in Rotated Sorted Array using Binary Search

nums = [17,18,19,20, 1, 3, 4, 5, 6, 7, 8, 10, 11, 12, 13, 14, 16]
target = 1

def search(nums, target):
    l, h = 0, len(nums) - 1   # low and high pointers
    
    while l <= h:
        mid = (l + h) // 2   # find middle index
        
        # Case 1: target found
        if nums[mid] == target:
            return mid
        
        # Case 2: Right half is sorted
        elif nums[mid] <= nums[h]:
            # Check if target lies in right sorted half
            if nums[mid] < target <= nums[h]:
                l = mid + 1   # move to right side
            else:
                h = mid - 1   # move to left side
        
        # Case 3: Left half is sorted
        else:  # nums[mid] >= nums[l]
            # Check if target lies in left sorted half
            if nums[l] <= target < nums[mid]:
                h = mid - 1   # move to left side
            else:
                l = mid + 1   # move to right side
    
    return -1   # target not found


# Function call
result = search(nums, target)
print("Index:", result)
```

```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        # check empty list
        if not strs:
            return ''
        
        prefix = strs[0]
        
        for word in strs[1:]:
            while not word.startswith(prefix):
                prefix = prefix[:-1]
                
                if prefix == '':
                    return ''
        
        return prefix
obj = Solution()

strs = ["flower", "flow", "flight"]

result = obj.longestCommonPrefix(strs)

print("Longest Common Prefix:", result)
```

```python
# Remove duplicates from sorted array
class Solution(object):
    def removeDuplicate(self, nums):
        if len(nums)==0: # empty list check
            return 0
        i=0 # pointer for unique elements
        for j in range(1,len(nums)): # traverse array
            if nums[j]!=nums[i]: # found new unique element
                i+=1 # move pointer
                nums[i]=nums[j] # place unique element
        return i+1 # count of unique elements

# Function call
obj=Solution() # create object
nums=[1,1,2,2,3,4,4,5] # input array
k=obj.removeDuplicate(nums) # call function
print("Unique count:",k) # print count
print("Updated array:",nums[:k]) # print unique elements
```

```python
# Remove all occurrences of val from array
class Solution(object):
    def removeElement(self,nums,val):
        k=0 # pointer for valid elements
        for i in range(len(nums)): # loop through array
            if nums[i]!=val: # keep elements not equal to val
                nums[k]=nums[i] # move element forward
                k+=1 # increment pointer
        return k # number of remaining elements

# Function call
obj=Solution() # create object
nums=[3,2,2,3,4,3,5] # input array
val=3 # value to remove
k=obj.removeElement(nums,val) # call function
print("Remaining count:",k) # print count
print("Updated array:",nums[:k]) # print result
```
