```python
# 📌 Program: Find Sum of Multiple Numbers using *args
def sum(*nums):
    total = 0
    for i in nums:
        total = total + i
    print(total)

sum(1, 2, 3, 4, 5, 6, 7)
```

```python
# Shallow copy (for non-nested list)
lst = [1, 2, 3, 4]

lst2 = lst.copy()   # creates a new separate list (shallow copy)

print(lst)   # original list
print(lst2)  # copied list

lst[0] = 0   # modify original list
print("lst1:", lst)

lst2[0] = 6  # modify copied list
print("lst2:", lst2)
```

```python
# Shallow copy with nested list
lst = [[1, 2, 3, 4], [5, 6, 7, 8]]

lst2 = lst.copy()   # creates a new outer list, but inner lists are still shared (same reference)

print(lst[0][3])   # access element 4
print(lst[1][3])   # access element 8

lst2[1][0] = 3     # modify inner list through copied list

print(lst2)        # change will be visible here
print(lst)         # original list also changes (same inner list reference)
```

```python
#abstract methed
from abc import ABC ,abstractmethod
class Vehicle(ABC):
    def fuel(self):
        print("Fuel is needed") 
    @abstractmethod
    def speed(self):
        pass
class Bike(Vehicle):
    def speed(self):
        print("bike speed is 80 km/h")
obj=Bike()
obj.fuel()
obj.speed()
```

```python
#decorator

def decorator(add):
    def wrapper(a,b):
        print('sum')
        return add(a,b)
    return wrapper

@decorator
def add(a,b):
    return a+b
print(add(1,3))
```

```python
# 📌 Program: Sum of All Prime Numbers Between 1 and N
print("write a program that will find the sum of all the prime numbers between 1 and N")

N = int(input("Enter N: "))

# Function to check prime number
def is_prime(num):
    if num <= 1:
        return False  # 0 and 1 are not prime numbers
    
    for i in range(2, num):
        if num % i == 0:
            return False  # Not prime if divisible
    
    return True  # Prime if no divisors found

total = 0
prime_number = []

# Loop through numbers from 2 to N
for i in range(2, N + 1):
    if is_prime(i):
        prime_number.append(i)  # Store prime number
        total += i              # Add to total sum

# Output results
print("Prime numbers:", prime_number)
print("Sum of prime numbers:", total)
```

```python
# 📌 Program: Check Palindrome String
nums = 'madam'

def palindrome(nums):
    f = 0
    n = len(nums) - 1
    l = n

    while f <= l:
        if nums[f] != nums[l]:
            return False   # Not a palindrome
        else:
            f = f + 1
            l = l - 1

    return True   # Palindrome

palindrome(nums)
```

```python
# 📌 Program: Fibonacci Series
def fib(n):
    print("Series")
    a, b = 0, 1

    print(a)   # First term of Fibonacci series

    for i in range(n):
        sum = a + b
        print(sum)   # Print next Fibonacci number
        a = b
        b = sum

    print("Sum of the series:", sum)

fib(5)
```

```python
# 📌 Program: FizzBuzz Problem
def fizzbuzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")   # Divisible by both 3 and 5
        elif i % 3 == 0:
            print('Fizz')       # Divisible by 3 only
        elif i % 5 == 0:
            print('Buzz')       # Divisible by 5 only
        else:
            print(i)            # Not divisible by 3 or 5

fizzbuzz(10)
```

```python
# 📌 Program: String Compression (Alternative Logic)
s = 'aabbccddeeee'
result = ''
count = 1

for i in range(1, len(s) + 1):
    if i < len(s) and s[i] == s[i - 1]:
        count += 1   # Same character → increase count
    else:
        result += s[i - 1] + str(count)   # Store character with its count
        count = 1   # Reset count for next character group

print(result)
```

```python
# 📌 Program: insert(), append() and direct assignment in list

# insert() → adds element at specific index (shifts elements)
lst = [1, 2, 3, 4]
lst.insert(0, 0)
print(lst)   # [0, 1, 2, 3, 4]

# append() → adds element at end of list
lst.append(5)
print(lst)   # [0, 1, 2, 3, 4, 5]

# direct assignment → replaces value at index
lst[0] = 6
print(lst)   # [6, 1, 2, 3, 4, 5]


# ---------------- SECOND EXAMPLE ----------------

lst = [1, 2, 3, 4]

lst.insert(1, 0)   # insert 0 at index 1
print(lst)         # [1, 0, 2, 3, 4]

lst.append(5)
print(lst)         # [1, 0, 2, 3, 4, 5]

lst[0] = 1
print(lst)         # [1, 0, 2, 3, 4, 5]
```

```python
# 📌 Program: Find Minimum and Maximum in a List

arr = [12, 0, 23, 4343, 43, 45]

min = arr[0]
max = arr[0]

for i in arr:
    if i < min:
        min = i   # update minimum value
    if i > max:
        max = i   # update maximum value

print('Minimum', min)   # 0 4343
print('Maximum',max) 
```

```python
# 📌 Program: all() and any() with Empty List

# all() with empty list
print(all([]))     # True

# Reason:
# all() → checks "are all elements True?"
# Empty list has no elements
# So there is no False element found → vacuously True


# any() with empty list
print(any([]))     # False

# Reason:
# any() → checks "is at least one element True?"
# Empty list has no elements
# So no True value exists → False
```

```python
# 📌 Program: List Reference Behavior (is vs ==)

a = [1, 2, 3, 4]
b = a   # both refer to same list object

b.append(5)
a.append(6)

print(a)   # [1, 2, 3, 4, 5, 6]
print(b)   # [1, 2, 3, 4, 5, 6]

print(a is b)   # True → same object in memory
print(a == b)   # True → values are equal
```

```python
# 📌 Program: Palindrome Check

text = 'madam'

def is_palindrome(text):
    start = 0
    end = len(text) - 1

    while start <= end:
        if text[start] != text[end]:
            return False   # not a palindrome
        start += 1
        end -= 1

    return True   # palindrome

print(is_palindrome(text))
```
