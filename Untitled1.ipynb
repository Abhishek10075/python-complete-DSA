```python
#Find the factorial of number using the recrussion
def fact(num):
    if num<0:
        return -1
    if num==0 or num==1:
        return 1
    else:
        return fact(num-1)*num
print(fact(5))
```

```python
nums = [1,1,1,2,3,5,6,7,7,9,12,13]
target = 14

def lowerbound(nums, target):
    l = 0
    h = len(nums) - 1
    lb = len(nums)  # default value

    while l <= h:
        mid = (l + h) // 2

        if nums[mid] >= target:
            lb = mid
            h = mid - 1
        else:
            l = mid + 1

    return lb

print(lowerbound(nums, target))
```

```python
nums = [1,1,1,2,3,5,6,7,7,9,12,13]
target = 1

def upperbound(nums, target):
    l = 0
    h = len(nums) - 1
    ub = len(nums)  # default value

    while l <= h:
        mid = (l + h) // 2

        if nums[mid] > target:
            ub = mid
            h = mid - 1
        else:
            l = mid + 1

    return ub

print(upperbound(nums, target))
```
